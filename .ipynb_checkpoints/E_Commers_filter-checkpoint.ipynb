# combinare linkedIn files

```python
import json
import glob
import os

os.chdir("C:/Users/iurie/E_Commers/Link")
# --- Configurari ---

# Directorul unde se afla fisierele JSON sursa
# (foloseste "." pentru directorul curent)
director_sursa = "." 

# Modelul fisierelor de intrare (bazat pe imaginea ta)
model_fisier = "dataset_linkedin-jobs-scraper_*.json"

# Numele fisierului final (curat si combinat)
fisier_output = "rezultat_final_unic.json"

# Cheia dupa care se face deduplicarea
cheie_unica = "jobUrl"

# --------------------

# Lista care va tine toate datele combinate (inainte de deduplicare)
lista_combinata = []

# Cautam fisierele
cale_cautare = os.path.join(director_sursa, model_fisier)
fisiere_de_intrare = glob.glob(cale_cautare)

# Ne asiguram ca nu procesam si fisierul de output, daca exista deja
cale_output_absoluta = os.path.abspath(os.path.join(director_sursa, fisier_output))
fisiere_de_intrare = [f for f in fisiere_de_intrare if os.path.abspath(f) != cale_output_absoluta]

if not fisiere_de_intrare:
    print(f"Nu am gasit fisiere care sa corespunda modelului '{model_fisier}' in '{director_sursa}'.")
else:
    print(f"Am gasit {len(fisiere_de_intrare)} fisiere de combinat...")

# --- Pasul 1: Combinarea tuturor fisierelor ---
numar_total_inregistrari = 0
for fisier in fisiere_de_intrare:
    print(f"Procesez: {fisier}")
    try:
        with open(fisier, 'r', encoding='utf-8') as f:
            data = json.load(f)
            
            # Presupunem ca majoritatea fisierelor contin o lista
            if isinstance(data, list):
                lista_combinata.extend(data)
                numar_total_inregistrari += len(data)
            # Daca e un singur obiect, il adaugam si pe el
            elif isinstance(data, dict):
                lista_combinata.append(data)
                numar_total_inregistrari += 1

    except json.JSONDecodeError:
        print(f"  Avertisment: Fisierul {fisier} este gol sau nu este un JSON valid. Va fi ignorat.")
    except Exception as e:
        print(f"  Eroare la citirea fisierului {fisier}: {e}")

print(f"\n--- Combinare finalizata ---")
print(f"Am gasit {numar_total_inregistrari} inregistrari in total (inainte de deduplicare).")


# --- Pasul 2: Deduplicarea ---
lista_finala_unica = []
elemente_vazute = set()
inregistrari_fara_cheie = 0
inregistrari_nevalide = 0

for item in lista_combinata:
    # Verificam daca e un obiect (dictionar)
    if not isinstance(item, dict):
        print(f"Avertisment: Am gasit o inregistrare care nu este un obiect valid. Se ignora.")
        inregistrari_nevalide += 1
        continue
    
    # Extragem valoarea cheii unice
    valoare_cheie = item.get(cheie_unica)
    
    # Daca are cheia si valoarea nu e nula
    if valoare_cheie:
        # Verificam daca am mai vazut-o
        if valoare_cheie not in elemente_vazute:
            # Daca e noua, o adaugam in lista finala si in set
            lista_finala_unica.append(item)
            elemente_vazute.add(valoare_cheie)
        # else: este un duplicat, deci il ignoram
            
    else:
        # Daca nu are cheia (ex: item[cheie_unica] e None sau nu exista)
        # O adaugam oricum, deoarece nu o putem compara
        lista_finala_unica.append(item)
        inregistrari_fara_cheie += 1

print(f"\n--- Deduplicare finalizata ---")
numar_duplicate = numar_total_inregistrari - len(lista_finala_unica) - inregistrari_nevalide
print(f"Am eliminat {numar_duplicate} duplicate.")
if inregistrari_fara_cheie > 0:
    print(f"Am pastrat {inregistrari_fara_cheie} inregistrari care nu aveau cheia '{cheie_unica}'.")
if inregistrari_nevalide > 0:
    print(f"Am ignorat {inregistrari_nevalide} inregistrari care nu erau obiecte JSON valide.")


# --- Pasul 3: Salvarea rezultatului ---
try:
    cale_output_finala = os.path.join(director_sursa, fisier_output)
    with open(cale_output_finala, 'w', encoding='utf-8') as f_out:
        # 'indent=4' face fisierul JSON usor de citit (pretty-print)
        # 'ensure_ascii=False' pastreaza caracterele speciale (diacritice)
        json.dump(lista_finala_unica, f_out, indent=4, ensure_ascii=False)
    
    print(f"\nSucces! 🚀 Au fost salvate {len(lista_finala_unica)} inregistrari unice in {fisier_output}.")
except Exception as e:
    print(f"\nEroare la scrierea fisierului de output: {e}")


print(f"Directorul curent: {os.getcwd()}")
calea_noua = "C:/Users/iurie/E_Commers"

try:
    os.chdir(calea_noua)

    print(f"Noul director: {os.getcwd()}")
except FileNotFoundError:
    print(f"Eroare: Directorul '{calea_noua}' nu a fost găsit.")
except Exception as e:
    print(f"A apărut o eroare: {e}")
```

# remove dublicates 

```python
import json
import glob
import os

# --- Configurarea Sarcinilor de Deduplicare ---
# Fiecare dictionar din aceasta lista defineste o sarcina separata.
# Scriptul va crea un fisier de iesire pentru fiecare sarcina.
#
# Poti modifica:
#  - 'model_intrare': Cum cauta fisierele (ex: "nume_fisier_*.json")
#  - 'fisier_iesire': Numele fisierului curat care va fi creat
#  - 'cheie_unica': Campul dupa care se sterg duplicatele

config_sarcini = [
    {
        "nume": "LinkedIn",
        "model_intrare": "dataset_linkedin_e_com*.json", 
        "fisier_iesire": "linkedin_unic.json",
        "cheie_unica": "jobUrl"
    },
    {
        "nume": "Glassdoor",
        "model_intrare": "dataset_glassdoor_e_com*.json", 
        "fisier_iesire": "glassdoor_unic.json",
        "cheie_unica": "apply_url"
    },
    {
        "nume": "Stepstone",
        "model_intrare": "dataset_stepstone_e_com*.json", 
        "fisier_iesire": "stepstone_unic.json",
        "cheie_unica": "id"
    }
]

# Directorul unde se afla fisierele JSON sursa
director_sursa = "." 

print("--- Incep procesarea separata a surselor ---")

# --- Functie ajutatoare pentru procesarea unei singure sarcini ---
def proceseaza_sarcina(sarcina, director_sursa):
    nume_sursa = sarcina["nume"]
    model_intrare = sarcina["model_intrare"]
    fisier_iesire = sarcina["fisier_iesire"]
    cheie_unica = sarcina["cheie_unica"]
    
    print(f"\n--- Procesare Sarcina: {nume_sursa} ---")
    print(f"  Cheie unica: '{cheie_unica}'")
    
    # Cautam fisierele
    cale_cautare = os.path.join(director_sursa, model_intrare)
    fisiere_de_intrare = glob.glob(cale_cautare)
    
    # Ne asiguram ca nu procesam si fisierul de output
    cale_output_absoluta = os.path.abspath(os.path.join(director_sursa, fisier_iesire))
    fisiere_de_intrare = [f for f in fisiere_de_intrare if os.path.abspath(f) != cale_output_absoluta]

    if not fisiere_de_intrare:
        print(f"  Nu am gasit fisiere pentru modelul '{model_intrare}'. Se ignora.")
        return

    print(f"  Am gasit {len(fisiere_de_intrare)} fisiere de procesat.")
    
    # --- Pas 1: Combinarea datelor (doar pentru fisierele acestei sarcini) ---
    lista_combinata_local = []
    numar_total_local = 0
    for fisier in fisiere_de_intrare:
        try:
            with open(fisier, 'r', encoding='utf-8') as f:
                data = json.load(f)
                if isinstance(data, list):
                    lista_combinata_local.extend(data)
                    numar_total_local += len(data)
                elif isinstance(data, dict):
                    lista_combinata_local.append(data)
                    numar_total_local += 1
        except json.JSONDecodeError:
            print(f"    Avertisment: Fisierul {fisier} este gol sau nu este un JSON valid.")
        except Exception as e:
            print(f"    Eroare la citirea fisierului {fisier}: {e}")
            
    if not lista_combinata_local:
        print("  Nu s-au gasit date valide in fisiere.")
        return

    print(f"  Am combinat {numar_total_local} inregistrari totale.")

    # --- Pasul 2: Deduplicarea ---
    lista_unica_local = []
    elemente_vazute = set()
    inregistrari_fara_cheie = 0
    inregistrari_nevalide = 0

    for item in lista_combinata_local:
        if not isinstance(item, dict):
            inregistrari_nevalide += 1
            continue
        
        valoare_cheie = item.get(cheie_unica)
        
        if valoare_cheie:
            if valoare_cheie not in elemente_vazute:
                lista_unica_local.append(item)
                elemente_vazute.add(valoare_cheie)
            # else: duplicat, ignoram
        else:
            lista_unica_local.append(item)
            inregistrari_fara_cheie += 1

    numar_duplicate = numar_total_local - len(lista_unica_local) - inregistrari_nevalide
    
    print("  Deduplicare finalizata:")
    print(f"    - {len(lista_unica_local)} inregistrari unice pastrate.")
    print(f"    - {numar_duplicate} duplicate eliminate.")
    if inregistrari_fara_cheie > 0:
        print(f"    - {inregistrari_fara_cheie} inregistrari pastrate fara cheie unica.")

    # --- Pasul 3: Salvarea fisierului unic pentru aceasta sarcina ---
    try:
        cale_output_finala = os.path.join(director_sursa, fisier_iesire)
        with open(cale_output_finala, 'w', encoding='utf-8') as f_out:
            json.dump(lista_unica_local, f_out, indent=4, ensure_ascii=False)
        
        print(f"  Succes! 🚀 Rezultatul a fost salvat in: {fisier_iesire}")
    except Exception as e:
        print(f"  Eroare la scrierea fisierului de output {fisier_iesire}: {e}")

# --- Executam fiecare sarcina din configuratie ---
for sarcina in config_sarcini:
    proceseaza_sarcina(sarcina, director_sursa)

print("\n--- Toate sarcinile au fost finalizate ---")
```

# Formatare comuna + joburi cu vechime 6 luni

```python
import json
import os
from datetime import datetime, date, timedelta
from dateutil.relativedelta import relativedelta
import re

# --- AVERTISMENT IMPORTANT LEGAT DE DATE ---
#
# Aceasta functie incearca sa parseze datele. Am inclus formate comune
# (ex: "2025-10-30T10:00:00Z", "2025-10-30", "30/10/2025", "Oct 30, 2025").
#
# DACA DATELE TALE SUNT IN ALT FORMAT (ex: "3 days ago", "ieri"),
# aceasta functie VA ESUA si acele joburi nu vor fi incluse.
# Va trebui sa ajustezi logica de parsare de mai jos.
#
# -----------------------------------------------
def parse_date(date_str):
    """
    Extrage data (primele 10 caractere) si o transforma 
    intr-un obiect 'date'.
    """
    if not date_str or not isinstance(date_str, str) or len(date_str) < 10:
        return None
    
    # Extragem doar portiunea YYYY-MM-DD
    date_only_str = date_str[:10]
    
    try:
        # Parsam folosind formatul exact
        dt = datetime.strptime(date_only_str, "%Y-%m-%d")
        return dt.date()
    except (ValueError, TypeError):
        # Acest lucru se poate intampla daca primele 10
        # caractere nu sunt "YYYY-MM-DD"
        print(f"  Avertisment: Nu am putut parsa data: '{date_str}'")
        return None

# --- FUNCTII DE NORMALIZARE (MAPARE) ---

def normalize_linkedin(job):
    """Transforma un job LinkedIn in formatul comun."""
    
    # Ambiție: Extragerea automata de skill-uri
    # Pentru "Skills" din "description", ar fi nevoie de NLP (procesare
    # complexa a limbajului). Deocamdata, il lasam gol (None).
    skills = None 
    
    common_job = {
        "Title": job.get("title"),
        "Location": job.get("location"),
        "PublishedDate": job.get("publishedAt"),
        "CompanyName": job.get("companyName"),
        "Description": job.get("description"),
        "Salary": job.get("salary"),
        "WorkSector": job.get("sector"),
        "WorkType": job.get("workType"),
        "Experience": job.get("experienceLevel"),
        "Skills": skills, # Setat la None, vezi nota de mai sus
        "JobType": job.get("contractType"),
        "Source": "linkedin"
    }
    return common_job

def normalize_glassdoor(job):
    """Transforma un job Glassdoor in formatul comun."""
    
    # Combinam "country" si "location"
    loc_parts = [job.get("country"), job.get("location")]
    # Filtram partile care nu sunt None si le unim
    location = " ".join(part for part in loc_parts if part)

    common_job = {
        "Title": job.get("title"),
        "Location": location or None, # Daca ambele sunt goale, punem None
        "PublishedDate": job.get("posted_date"),
        "CompanyName": job.get("company_name"),
        "Description": job.get("skills"), # Conform cerintei
        "Salary": job.get("salary_range"),
        "WorkSector": "N/A", # Hardcodat
        "WorkType": job.get("category"),
        "Experience": job.get("experience"),
        "Skills": job.get("skills"), # Conform cerintei
        "JobType": job.get("job_type"),
        "Source": "glassdoor"
    }
    return common_job

def normalize_stepstone(job):
    """Transforma un job Stepstone in formatul comun."""
    
    common_job = {
        "Title": job.get("title"),
        "Location": job.get("location"),
        "PublishedDate": job.get("datePosted"),
        "CompanyName": job.get("companyName"),
        "Description": job.get("textSnippet"),
        "Salary": job.get("salary"),
        "WorkSector": "N/A", # Hardcodat
        "WorkType": "N/A", # Hardcodat
        "Experience": "N/A", # Hardcodat
        "Skills": job.get("skills"),
        "JobType": job.get("job_type"),
        "Source": "stepstone" # Am corectat (era "glassdoor" in cerinta)
    }
    return common_job

# --- CONFIGURAREA SURSELOR ---

# Directorul unde se afla fisierele
director_sursa = "."

# Definim fisierele de intrare si functiile lor de normalizare
CONFIG_SURSE = [
    {
        "fisier": "linkedin_unic.json",
        "functie_norm": normalize_linkedin
    },
    {
        "fisier": "glassdoor_unic.json",
        "functie_norm": normalize_glassdoor
    },
    {
        "fisier": "stepstone_unic.json",
        "functie_norm": normalize_stepstone
    }
]

# Fisierul final
FISIER_OUTPUT_FINAL = "E_Commers_jobs.json"

# --- PROCESARE PRINCIPALA ---

print("--- Incep normalizarea si filtrarea joburilor ---")

# Pasul 1: Definirea perioadei de timp
# "Today()+1" - pentru a include si joburile postate astazi
ziua_de_referinta = date.today() + timedelta(days=1)
# Data limita (acum 6 luni)
data_limita = ziua_de_referinta - relativedelta(months=6)

print(f"Data de referinta (ziua de azi + 1): {ziua_de_referinta}")
print(f"Data limita (se pastreaza joburile >=): {data_limita}")

# Lista finala care va contine joburile normalizate SI filtrate
lista_finala_filtrata = []
total_procesate = 0
total_pastrate = 0
total_expirate = 0
total_fara_data = 0

# Pasul 2: Iteram prin fiecare fisier sursa
for sursa in CONFIG_SURSE:
    nume_fisier = sursa["fisier"]
    functie_norm = sursa["functie_norm"]
    cale_fisier = os.path.join(director_sursa, nume_fisier)
    
    print(f"\nProcesez fisierul: {nume_fisier}...")
    
    try:
        with open(cale_fisier, 'r', encoding='utf-8') as f:
            lista_joburi = json.load(f)
            
            if not isinstance(lista_joburi, list):
                print(f"  Eroare: Fisierul {nume_fisier} nu contine o lista JSON.")
                continue
                
            print(f"  Am gasit {len(lista_joburi)} joburi de procesat.")
            
            # Pasul 3: Normalizare si Filtrare pentru fiecare job
            for job_brut in lista_joburi:
                total_procesate += 1
                
                # 3a. Normalizare
                job_comun = functie_norm(job_brut)
                
                # 3b. Parsare data
                data_publicare_str = job_comun.get("PublishedDate")
                data_publicare_obj = parse_date(data_publicare_str)
                
                # 3c. Filtrare
                if data_publicare_obj:
                    if data_publicare_obj >= data_limita:
                        # Jobul este suficient de recent
                        lista_finala_filtrata.append(job_comun)
                        total_pastrate += 1
                    else:
                        # Jobul este prea vechi (expirat)
                        total_expirate += 1
                else:
                    # Jobul nu are o data de publicare valida
                    # Poti decide daca vrei sa le pastrezi sau nu.
                    # Momentan, le ignoram.
                    total_fara_data += 1

    except FileNotFoundError:
        print(f"  Eroare: Fisierul {nume_fisier} nu a fost gasit.")
    except json.JSONDecodeError:
        print(f"  Eroare: Fisierul {nume_fisier} nu este un JSON valid.")
    except Exception as e:
        print(f"  A aparut o eroare neasteptata la {nume_fisier}: {e}")

# Pasul 4: Salvarea rezultatului final
print("\n--- Finalizare Proces ---")
print(f"Total joburi procesate: {total_procesate}")
print(f"Joburi pastrate (in ultimele 6 luni): {total_pastrate}")
print(f"Joburi eliminate (prea vechi): {total_expirate}")
print(f"Joburi eliminate (fara data parsabila): {total_fara_data}")

try:
    cale_output_finala = os.path.join(director_sursa, FISIER_OUTPUT_FINAL)
    with open(cale_output_finala, 'w', encoding='utf-8') as f_out:
        json.dump(lista_finala_filtrata, f_out, indent=4, ensure_ascii=False)
    
    print(f"\nSucces! 🚀 Fisierul combinat si filtrat a fost salvat in: {FISIER_OUTPUT_FINAL}")
except Exception as e:
    print(f"\nEroare la scrierea fisierului de output: {e}")
```

# completare_skills

```python
import json
import os
import glob
import re

# --- Configurari Fisiere ---
SKILLS_FILE = "unique_skills.json"
GLASSDOOR_GLOB = "dataset_glassdoor_e_com*.json"
JOBS_FILE = "E_Commers_jobs.json"

# =====================================================================
# PARTEA 1: Actualizarea listei de skill-uri unice din Glassdoor
# =====================================================================

print(f"--- Partea 1: Actualizare {SKILLS_FILE} ---")

def load_existing_skills(filepath):
    """Incarca skill-urile existente dintr-un fisier JSON."""
    if not os.path.exists(filepath):
        print(f"Info: Fisierul {filepath} nu exista. Se va crea unul nou.")
        return set() # Returneaza un set gol
        
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            data = json.load(f)
            if isinstance(data, list):
                # Convertim lista intr-un set pentru deduplicare rapida
                return set(data)
            else:
                print(f"Avertisment: {filepath} nu contine o lista. Se incepe de la zero.")
                return set()
    except json.JSONDecodeError:
        print(f"Eroare: {filepath} este gol sau corupt. Se incepe de la zero.")
        return set()
    except Exception as e:
        print(f"Eroare la citirea {filepath}: {e}")
        return set()

# Incarcam skill-urile existente
current_skills_set = load_existing_skills(SKILLS_FILE)
print(f"Am incarcat {len(current_skills_set)} skill-uri unice existente.")

# Cautam fisierele Glassdoor
glassdoor_files = glob.glob(GLASSDOOR_GLOB)
if not glassdoor_files:
    print(f"Avertisment: Nu am gasit fisiere care sa corespunda modelului '{GLASSDOOR_GLOB}'")
else:
    print(f"Procesez {len(glassdoor_files)} fisiere Glassdoor pentru a extrage skill-uri...")

new_skills_count = 0
for f_path in glassdoor_files:
    try:
        with open(f_path, 'r', encoding='utf-8') as f:
            job_list = json.load(f)
            
            if not isinstance(job_list, list):
                # Daca fisierul nu contine o lista, il ignoram
                continue

            # Extragem skill-urile
            for job in job_list:
                job_skills = job.get("skills")
                
                if isinstance(job_skills, list):
                    # Daca e o lista de skill-uri
                    current_skills_set.update(job_skills)
                    new_skills_count += len(job_skills)
                elif isinstance(job_skills, str) and job_skills:
                    # Daca e un singur string (desi e mai putin probabil)
                    current_skills_set.add(job_skills)
                    new_skills_count += 1
                    
    except json.JSONDecodeError:
        print(f"  Avertisment: Fisierul {f_path} este gol sau nu este un JSON valid. Va fi ignorat.")
    except Exception as e:
        print(f"  Eroare la citirea fisierului {f_path}: {e}")

print(f"Am adaugat {new_skills_count} skill-uri (inainte de deduplicare) din Glassdoor.")

# Sortam lista finala pentru a fi usor de citit
unique_skills_list = sorted(list(current_skills_set))

print(f"Total skill-uri unice acum: {len(unique_skills_list)}")

# Salvam noua lista de skill-uri unice
try:
    with open(SKILLS_FILE, 'w', encoding='utf-8') as f:
        json.dump(unique_skills_list, f, indent=4, ensure_ascii=False)
    print(f"Succes! Fisierul {SKILLS_FILE} a fost actualizat.")
except Exception as e:
    print(f"Eroare la scrierea {SKILLS_FILE}: {e}")


# =====================================================================
# PARTEA 2: Completarea skill-urilor in E_Commers_jobs.json
# =====================================================================

print(f"\n--- Partea 2: Completare skill-uri in {JOBS_FILE} ---")

if not unique_skills_list:
    print("Lista de skill-uri este goala. Nu se poate continua cu Partea 2.")
    exit()

# Optimizare: Compilam expresiile regulate o singura data
# Acest dictionar va mapa skill-ul original (ex: "C++")
# la expresia regulata compilata (ex: re.compile(r'\bc\+\+\b', re.IGNORECASE))
skill_regex_map = {}
for skill in unique_skills_list:
    # re.escape() trateaza caractere speciale (ex: C++, Node.js)
    # \b asigura potrivirea cuvantului intreg (gaseste "Java", nu "Javascript")
    # re.IGNORECASE face cautarea insensibila la majuscule/minuscule
    try:
        regex = re.compile(r'\b' + re.escape(skill) + r'\b', re.IGNORECASE)
        skill_regex_map[skill] = regex
    except re.error as e:
        print(f"  Avertisment: Nu am putut compila regex pentru '{skill}': {e}")

print(f"Am compilat {len(skill_regex_map)} expresii regex pentru cautare.")

# Incarcam fisierul mare de joburi
try:
    with open(JOBS_FILE, 'r', encoding='utf-8') as f:
        e_commers_jobs = json.load(f)
except FileNotFoundError:
    print(f"Eroare: Fisierul {JOBS_FILE} nu a fost gasit. Opreste scriptul.")
    exit()
except Exception as e:
    print(f"Eroare la citirea {JOBS_FILE}: {e}")
    exit()

if not isinstance(e_commers_jobs, list):
    print(f"Eroare: {JOBS_FILE} nu contine o lista. Opreste scriptul.")
    exit()

print(f"Am incarcat {len(e_commers_jobs)} joburi din {JOBS_FILE}.")

jobs_updated_count = 0
# Iteram prin fiecare job si il modificam
for job in e_commers_jobs:
    # Procesam doar sursele specificate
    if job.get("Source") in ["linkedin", "stepstone"]:
        description = job.get("Description")
        
        # Daca nu are descriere, nu putem cauta
        if not description or not isinstance(description, str):
            job["Skills"] = [] # Setam o lista goala
            continue
            
        found_skills = set()
        
        # Cautam fiecare skill in descriere
        for skill_name, regex in skill_regex_map.items():
            if regex.search(description):
                # Adaugam skill-ul original (ex: "Python"), nu "python"
                found_skills.add(skill_name)
        
        # Actualizam campul "Skills" cu lista de skill-uri gasite
        job["Skills"] = list(found_skills)
        
        if found_skills:
            jobs_updated_count += 1

print(f"Am actualizat {jobs_updated_count} joburi (LinkedIn/Stepstone) cu skill-uri extrase.")

# Salvam fisierul modificat
try:
    with open(JOBS_FILE, 'w', encoding='utf-8') as f:
        json.dump(e_commers_jobs, f, indent=4, ensure_ascii=False)
    print(f"Succes! 🚀 Fisierul {JOBS_FILE} a fost suprascris cu datele actualizate.")
except Exception as e:
    print(f"Eroare la scrierea finala in {JOBS_FILE}: {e}")
```

# filtrare_job_title

```python
import json
import os

# --- Configurari Fisiere ---
JOBS_FILE = "E_Commers_jobs.json"
TITLES_FILE = "job_title.json"
OUTPUT_FILE = "E_Commers_jobs_filtrat_titlu.json"

print("--- Incep filtrarea joburilor dupa titlu ---")

# --- Pasul 1: Incarcarea listei de titluri de joburi ---
try:
    with open(TITLES_FILE, 'r', encoding='utf-8') as f:
        data = json.load(f)
        
        # Extragem lista de titluri din cheia "job title"
        job_titles_raw = data.get("job title")
        
        if not isinstance(job_titles_raw, list):
            print(f"Eroare: Fisierul {TITLES_FILE} nu are structura asteptata (ex: {{'job title': [...]}})")
            exit()
            
        # Convertim toate titlurile in litere mici (lower register)
        job_titles_lower = [str(title).lower() for title in job_titles_raw if title]
        
    print(f"Am incarcat {len(job_titles_lower)} titluri de joburi din {TITLES_FILE}.")

except FileNotFoundError:
    print(f"Eroare: Fisierul de titluri {TITLES_FILE} nu a fost gasit.")
    exit()
except json.JSONDecodeError:
    print(f"Eroare: Fisierul {TITLES_FILE} nu este un JSON valid.")
    exit()
except Exception as e:
    print(f"Eroare la citirea {TITLES_FILE}: {e}")
    exit()

if not job_titles_lower:
    print("Eroare: Nu am gasit niciun titlu de job in fisier. Operatiune oprita.")
    exit()

# --- Pasul 2: Incarcarea fisierului mare de joburi ---
try:
    with open(JOBS_FILE, 'r', encoding='utf-8') as f:
        e_commers_jobs = json.load(f)
        
    if not isinstance(e_commers_jobs, list):
        print(f"Eroare: Fisierul {JOBS_FILE} nu contine o lista JSON.")
        exit()
        
    print(f"Am incarcat {len(e_commers_jobs)} joburi din {JOBS_FILE} pentru filtrare.")
    
except FileNotFoundError:
    print(f"Eroare: Fisierul de joburi {JOBS_FILE} nu a fost gasit.")
    exit()
except Exception as e:
    print(f"Eroare la citirea {JOBS_FILE}: {e}")
    exit()

# --- Pasul 3: Filtrarea joburilor ---
filtered_jobs = []
jobs_discarded = 0

for job in e_commers_jobs:
    # Extragem titlul si descrierea
    title_str = job.get("Title")
    desc_str = job.get("Description")
    
    # Convertim in litere mici (lower register). 
    # Daca un camp e gol (None), il tratam ca string gol.
    title_lower = title_str.lower() if isinstance(title_str, str) else ""
    desc_lower = desc_str.lower() if isinstance(desc_str, str) else ""
    
    # Combinam titlul si descrierea intr-un singur text
    # pentru a cauta o singura data
    search_text = title_lower + " " + desc_lower
    
    match_found = False
    
    # Cautam daca VREUNUL din titlurile noastre
    # se regaseste in textul jobului
    for filter_title in job_titles_lower:
        if filter_title in search_text:
            match_found = True
            # Am gasit o potrivire, nu mai e nevoie
            # sa cautam alte titluri in acest job
            break 
            
    # Daca am gasit o potrivire, pastram jobul
    if match_found:
        filtered_jobs.append(job)
    else:
        jobs_discarded += 1

# --- Pasul 4: Salvarea rezultatului ---
print("\n--- Filtrare Finalizata ---")
print(f"Total joburi procesate: {len(e_commers_jobs)}")
print(f"Joburi pastrate (care se potrivesc): {len(filtered_jobs)}")
print(f"Joburi eliminate (fara potrivire): {jobs_discarded}")

try:
    with open(OUTPUT_FILE, 'w', encoding='utf-8') as f_out:
        json.dump(filtered_jobs, f_out, indent=4, ensure_ascii=False)
    
    print(f"\nSucces! 🚀 Fisierul filtrat a fost salvat in: {OUTPUT_FILE}")
except Exception as e:
    print(f"\nEroare la scrierea fisierului de output: {e}")
```

# v2

```python
import json
import os

# --- Configurari Fisiere ---
JOBS_FILE = "E_Commers_jobs.json"
TITLES_FILE = "job_title.json"
OUTPUT_FILE = "E_Commers_jobs_filtrat_titlu_V2.json"

# --- CAMPURILE in care vom cauta cuvintele cheie ---
# Am adaugat 'WorkSector' si 'WorkType'
CAMPURI_DE_CAUTARE = ["Title", "Description", "WorkSector", "WorkType"]


print("--- Incep filtrarea V2 (cautare extinsa) ---")
print(f"Vom cauta in campurile: {CAMPURI_DE_CAUTARE}")

# --- Pasul 1: Incarcarea listei de titluri de joburi (cuvinte cheie) ---
try:
    with open(TITLES_FILE, 'r', encoding='utf-8') as f:
        data = json.load(f)
        job_titles_raw = data.get("job title")
        
        if not isinstance(job_titles_raw, list):
            print(f"Eroare: Fisierul {TITLES_FILE} nu are structura asteptata.")
            exit()
            
        job_titles_lower = [str(title).lower() for title in job_titles_raw if title]
        
    print(f"Am incarcat {len(job_titles_lower)} cuvinte cheie din {TITLES_FILE}.")

except FileNotFoundError:
    print(f"Eroare: Fisierul de titluri {TITLES_FILE} nu a fost gasit.")
    exit()
except Exception as e:
    print(f"Eroare la citirea {TITLES_FILE}: {e}")
    exit()

if not job_titles_lower:
    print("Eroare: Nu am gasit niciun cuvant cheie in fisier.")
    exit()

# --- Pasul 2: Incarcarea fisierului mare de joburi ---
try:
    with open(JOBS_FILE, 'r', encoding='utf-8') as f:
        e_commers_jobs = json.load(f)
        
    if not isinstance(e_commers_jobs, list):
        print(f"Eroare: Fisierul {JOBS_FILE} nu contine o lista JSON.")
        exit()
        
    print(f"Am incarcat {len(e_commers_jobs)} joburi din {JOBS_FILE} pentru filtrare.")
    
except FileNotFoundError:
    print(f"Eroare: Fisierul de joburi {JOBS_FILE} nu a fost gasit.")
    exit()
except Exception as e:
    print(f"Eroare la citirea {JOBS_FILE}: {e}")
    exit()

# --- Pasul 3: Filtrarea joburilor (logica noua) ---
filtered_jobs = []
jobs_discarded = 0

for job in e_commers_jobs:
    
    # Combinam textul din TOATE campurile relevante
    search_text = ""
    for camp in CAMPURI_DE_CAUTARE:
        valoare_camp = job.get(camp)
        if isinstance(valoare_camp, str):
            search_text += " " + valoare_camp.lower()
        elif isinstance(valoare_camp, list):
            # Daca un camp e lista (ex: Skills), unim elementele
            search_text += " " + " ".join(valoare_camp).lower()
            
    match_found = False
    
    for filter_title in job_titles_lower:
        if filter_title in search_text:
            match_found = True
            break 
            
    if match_found:
        filtered_jobs.append(job)
    else:
        jobs_discarded += 1

# --- Pasul 4: Salvarea rezultatului ---
print("\n--- Filtrare V2 Finalizata ---")
print(f"Total joburi procesate: {len(e_commers_jobs)}")
print(f"Joburi pastrate (care se potrivesc): {len(filtered_jobs)}")
print(f"Joburi eliminate (fara potrivire): {jobs_discarded}")

try:
    with open(OUTPUT_FILE, 'w', encoding='utf-8') as f_out:
        json.dump(filtered_jobs, f_out, indent=4, ensure_ascii=False)
    
    print(f"\nSucces! 🚀 Fisierul filtrat (V2) a fost salvat in: {OUTPUT_FILE}")
except Exception as e:
    print(f"\nEroare la scrierea fisierului de output: {e}")
```

# analiza_skills

```python
import json
import os

# --- Configurari ---
FILE_TO_CLEAN = "E_Commers_jobs_filtrat_titlu_V2.json"

# Definim regulile de normalizare
# Cheia este varianta scrisa cu litera mica,
# Valoarea este forma finala (corecta)
map_normalizare = {
    "e-commerce": "E-Commerce",
    "digital marketing": "Digital Marketing",
    "customer service" : "Customer Service"
}

print(f"--- Incep normalizarea skill-urilor in {FILE_TO_CLEAN} ---")

# --- Incarcam fisierul ---
try:
    with open(FILE_TO_CLEAN, 'r', encoding='utf-8') as f:
        data = json.load(f)
except FileNotFoundError:
    print(f"Eroare: Fisierul {FILE_TO_CLEAN} nu a fost gasit.")
    exit()
except Exception as e:
    print(f"Eroare la citirea fisierului: {e}")
    exit()

if not isinstance(data, list):
    print("Eroare: Fisierul nu contine o lista JSON.")
    exit()

print(f"Am incarcat {len(data)} joburi. Incep procesarea...")

# --- Procesam fiecare job ---
jobs_processed = 0
skills_normalized_count = 0

for job in data:
    skills_list = job.get("Skills")
    
    # Verificam daca are o lista de skill-uri
    if isinstance(skills_list, list) and skills_list:
        jobs_processed += 1
        new_skills_set = set() # Folosim un set pentru a elimina duplicatele
        needs_normalization = False

        for skill in skills_list:
            skill_lower = skill.lower()
            
            # Verificam daca skill-ul (in format mic)
            # este in harta noastra de normalizare
            if skill_lower in map_normalizare:
                # Daca da, adaugam forma corecta
                new_skills_set.add(map_normalizare[skill_lower])
                needs_normalization = True
            else:
                # Altfel, adaugam skill-ul original
                new_skills_set.add(skill)
        
        if needs_normalization:
            # Actualizam lista de skill-uri a jobului
            job["Skills"] = list(new_skills_set)
            skills_normalized_count += 1

print(f"Procesare finalizata.")
print(f"Am verificat {jobs_processed} joburi care aveau skill-uri.")
print(f"Am normalizat skill-uri pentru {skills_normalized_count} joburi.")

# --- Salvam fisierul modificat ---
try:
    with open(FILE_TO_CLEAN, 'w', encoding='utf-8') as f_out:
        json.dump(data, f_out, indent=4, ensure_ascii=False)
    
    print(f"\nSucces! 🚀 Fisierul {FILE_TO_CLEAN} a fost actualizat si suprascris.")
except Exception as e:
    print(f"\nEroare la scrierea fisierului de output: {e}")
```

```python
import json
from collections import Counter

# --- Configurari ---

# Acesta trebuie sa fie fisierul FILTRAT (cel cu 1581 de joburi)
FILE_TO_ANALYZE = "E_Commers_jobs_filtrat_titlu_V2.json"

# Numarul de skill-uri de top pe care sa le afiseze
TOP_N_SKILLS = 30

# --- Proces ---

print(f"--- Incep analiza skill-urilor din {FILE_TO_ANALYZE} ---")

# Counter este un dictionar specializat pentru numarare
skill_counter = Counter()
jobs_with_skills = 0
total_jobs = 0

try:
    with open(FILE_TO_ANALYZE, 'r', encoding='utf-8') as f:
        data = json.load(f)

    if not isinstance(data, list):
        print(f"Eroare: {FILE_TO_ANALYZE} nu contine o lista.")
        exit()
        
    total_jobs = len(data)
    
    # Iteram prin joburile FILTRATE
    for job in data:
        skills = job.get("Skills")
        
        # Verificam daca jobul are skill-uri
        if isinstance(skills, list) and skills:
            jobs_with_skills += 1
            # Adaugam lista de skill-uri la counter
            skill_counter.update(skills)

    if total_jobs == 0:
        print("Fisierul de joburi este gol. Nu se poate analiza.")
        exit()
        
    if not skill_counter:
        print("Nu am gasit niciun skill in campurile 'Skills' ale joburilor.")
        print("Asigura-te ca ai rulat scriptul 'completare_skills.py' inainte.")
        exit()

    # --- Afisarea Rezultatelor ---
    print(f"\n--- Rezultate Analiza ---")
    print(f"Total joburi analizate: {total_jobs}")
    print(f"Joburi care au cel putin un skill identificat: {jobs_with_skills} ({jobs_with_skills / total_jobs:.1%})")
    
    print(f"\n--- Top {TOP_N_SKILLS} Cele Mai Cerute Skill-uri ---")
    
    # .most_common(N) returneaza o lista de tupluri: [('Skill', numar), ...]
    for skill, count in skill_counter.most_common(TOP_N_SKILLS):
        # Calculam procentajul
        percentage = (count / jobs_with_skills) * 100
        print(f"{skill:<30} | Apare in {count:>5} joburi ({percentage:>5.1f}%)")

except FileNotFoundError:
    print(f"Eroare: Fisierul {FILE_TO_ANALYZE} nu a fost gasit.")
except Exception as e:
    print(f"A aparut o eroare: {e}")
```

# analiza_locatii_incomplete

```python
import json
import os
from collections import defaultdict

INPUT_FILE = "E_Commers_jobs_filtrat_titlu_V2.json"
OUTPUT_FILE = "raport_locatii_incomplete.json"

print(f"--- Incep analiza formatelor de locatie din {INPUT_FILE} ---")

# Vom folosi seturi pentru a stoca doar valorile unice
formate = {
    1: set(),
    2: set(),
    3: set(),
    'altele': set()
}

try:
    with open(INPUT_FILE, 'r', encoding='utf-8') as f:
        data = json.load(f)
except FileNotFoundError:
    print(f"Eroare: Fisierul {INPUT_FILE} nu a fost gasit.")
    exit()

for job in data:
    loc_str = job.get("Location")
    if isinstance(loc_str, str) and loc_str.strip():
        parts = [p.strip() for p in loc_str.split(",")]
        num_parts = len(parts)
        
        if num_parts == 1:
            formate[1].add(loc_str)
        elif num_parts == 2:
            formate[2].add(loc_str)
        elif num_parts == 3:
            formate[3].add(loc_str)
        else:
            formate['altele'].add(loc_str)

print("\n--- Sumar Formate Gasite ---")
print(f"Locatii cu 3 parti (ex: Oras, Regiune, Tara): {len(formate[3])} unice")
print(f"Locatii cu 2 parti (ex: Oras, Tara):         {len(formate[2])} unice")
print(f"Locatii cu 1 parte (ex: Tara):            {len(formate[1])} unice")
print(f"Locatii cu alt format (ex: 4+ parti):     {len(formate['altele'])} unice")

# Pregatim fisierul de output
raport = {
    "locatii_cu_1_parte (de revizuit)": sorted(list(formate[1])),
    "locatii_cu_2_parti (de revizuit)": sorted(list(formate[2])),
    "locatii_cu_3_parti (corecte)": sorted(list(formate[3]))
}

# --- Salvam raportul ---
try:
    with open(OUTPUT_FILE, 'w', encoding='utf-8') as f_out:
        json.dump(raport, f_out, indent=4, ensure_ascii=False)
    print(f"\nSucces! Raportul locatiilor a fost salvat in: {OUTPUT_FILE}")
except Exception as e:
    print(f"\nEroare la scrierea raportului: {e}")
```

# divizare_locatii

```python
import json
import os

# --- Configurari ---
INPUT_FILE = "E_Commers_jobs_filtrat_titlu_V2.json"
OUTPUT_FILE = "E_Commers_jobs_cu_locatii_divizate.json"

print(f"--- Incep divizarea locatiilor din {INPUT_FILE} ---")

# --- Incarcam fisierul ---
try:
    with open(INPUT_FILE, 'r', encoding='utf-8') as f:
        data = json.load(f)
except FileNotFoundError:
    print(f"Eroare: Fisierul {INPUT_FILE} nu a fost gasit.")
    exit()
except Exception as e:
    print(f"Eroare la citirea fisierului: {e}")
    exit()

if not isinstance(data, list):
    print("Eroare: Fisierul nu contine o lista JSON.")
    exit()

print(f"Am incarcat {len(data)} joburi. Incep procesarea...")

# --- Procesam fiecare job ---
locatii_procesate = 0
for job in data:
    # Extragem locatia originala
    loc_str = job.get("Location")
    
    # Initializam noile campuri cu None
    job["city"] = None
    job["region"] = None
    job["country"] = None

    # Daca locatia este un string valid, incercam sa o divizam
    if isinstance(loc_str, str) and loc_str.strip():
        locatii_procesate += 1
        
        # Curatam spatiile si divizam dupa virgula
        parts = [p.strip() for p in loc_str.split(",")]
        num_parts = len(parts)

        # Aplicam logica de divizare
        if num_parts == 1:
            # Format: "Germany"
            job["country"] = parts[0]
            
        elif num_parts == 2:
            # Format: "Hamburg, Germany"
            job["city"] = parts[0]
            job["country"] = parts[1]
            
        elif num_parts == 3:
            # Format: "Düsseldorf, North Rhine-Westphalia, Germany"
            job["city"] = parts[0]
            job["region"] = parts[1]
            job["country"] = parts[2]
            
        else:
            # Format necunoscut (ex: 4+ parti)
            # Il punem pe tot in "city" pentru a nu pierde data
            print(f"  Avertisment: Format locatie necunoscut: '{loc_str}'")
            job["city"] = loc_str

# --- Salvam fisierul modificat ---
print(f"\nProcesare finalizata. Am divizat {locatii_procesate} locatii.")

try:
    with open(OUTPUT_FILE, 'w', encoding='utf-8') as f_out:
        json.dump(data, f_out, indent=4, ensure_ascii=False)
    
    print(f"\nSucces! 🚀 Fisierul nou a fost salvat in: {OUTPUT_FILE}")
    print("Acest fisier poate fi incarcat direct in pandas/geopandas.")
    
except Exception as e:
    print(f"\nEroare la scrierea fisierului de output: {e}")
```

```python
import json
import os

# --- Configurari ---
# Sursa (fisierul filtrat)
INPUT_FILE = "E_Commers_jobs_filtrat_titlu_V2.json"
# Iesirea principala (va fi suprascris)
OUTPUT_FILE = "E_Commers_jobs_cu_locatii_divizate.json"
# Iesirea pentru joburile care necesita revizuire manuala
PARTIAL_LOCATIONS_FILE = "locatii_partiale_de_revizuit.json"

print(f"--- Incep procesarea avansata a locatiilor din {INPUT_FILE} ---")

# --- Incarcam fisierul sursa ---
try:
    with open(INPUT_FILE, 'r', encoding='utf-8') as f:
        data = json.load(f)
except FileNotFoundError:
    print(f"Eroare: Fisierul {INPUT_FILE} nu a fost gasit.")
    exit()
except Exception as e:
    print(f"Eroare la citirea fisierului: {e}")
    exit()

if not isinstance(data, list):
    print("Eroare: Fisierul nu contine o lista JSON.")
    exit()

print(f"Am incarcat {len(data)} joburi. Incep procesarea...")

# --- Initializam listele de iesire ---
final_jobs_list = [] # Pentru OUTPUT_FILE
partial_jobs_list = [] # Pentru PARTIAL_LOCATIONS_FILE

# --- Contoare pentru statistici ---
count_completate_berlin = 0
count_partiale_mutate = 0
count_corecte_pastrate = 0

# --- Etapa 1 & 2: Divizare, Reordonare și Filtrare ---
for job in data:
    # Cream un nou dictionar pentru a garanta ordinea campurilor
    new_job_ordered = {}
    
    # Valori temporare pentru logica de divizare
    city_val, region_val, country_val = None, None, None
    loc_str = None
    
    # Reconstruim jobul pentru a insera campurile
    for key, value in job.items():
        new_job_ordered[key] = value
        
        # Cand gasim "Location", inseram noile campuri imediat dupa
        if key == "Location":
            loc_str = value
            
            # --- Logica de divizare (de la scriptul anterior) ---
            if isinstance(loc_str, str) and loc_str.strip():
                parts = [p.strip() for p in loc_str.split(",")]
                num_parts = len(parts)

                if num_parts == 1:
                    country_val = parts[0]
                elif num_parts == 2:
                    city_val = parts[0]
                    country_val = parts[1]
                elif num_parts == 3:
                    city_val = parts[0]
                    region_val = parts[1]
                    country_val = parts[2]
                else:
                    city_val = loc_str # Fallback

            # Adaugam campurile divizate in noul dictionar
            new_job_ordered["city"] = city_val
            new_job_ordered["region"] = region_val
            new_job_ordered["country"] = country_val

    # --- Etapa 3: Logica Berlin / Partial ---
    
    # Verificam valorile pe care tocmai le-am extras
    
    if new_job_ordered["city"] is None and new_job_ordered["region"] is None:
        # 1. Ambele sunt goale (ex: "Germany") -> Completam cu "Berlin"
        new_job_ordered["city"] = "Berlin"
        new_job_ordered["region"] = "Berlin"
        final_jobs_list.append(new_job_ordered)
        count_completate_berlin += 1
        
    elif new_job_ordered["city"] is None or new_job_ordered["region"] is None:
        # 2. Doar una este goala (ex: "Hamburg, Germany") -> Mutam in fisierul partial
        partial_jobs_list.append(new_job_ordered)
        count_partiale_mutate += 1
        
    else:
        # 3. Totul e in regula (ex: "Düsseldorf, NRW, Germany") -> Pastram
        final_jobs_list.append(new_job_ordered)
        count_corecte_pastrate += 1

# --- Raport Final ---
print("\n--- Procesare Finalizata ---")
print(f"Total joburi procesate: {len(data)}")
print(f"Joburi corecte pastrate: {count_corecte_pastrate}")
print(f"Joburi completate cu 'Berlin': {count_completate_berlin}")
print(f"Joburi mutate (partiale): {count_partiale_mutate}")

# --- Salvarea fisierului principal ---
try:
    with open(OUTPUT_FILE, 'w', encoding='utf-8') as f_out:
        json.dump(final_jobs_list, f_out, indent=4, ensure_ascii=False)
    
    print(f"\nSucces! 🚀 Fisierul principal a fost suprascris in: {OUTPUT_FILE}")
    
except Exception as e:
    print(f"\nEroare la scrierea fisierului principal: {e}")

# --- Salvarea fisierului partial (doar daca exista) ---
if partial_jobs_list:
    try:
        with open(PARTIAL_LOCATIONS_FILE, 'w', encoding='utf-8') as f_out:
            json.dump(partial_jobs_list, f_out, indent=4, ensure_ascii=False)
        
        print(f"Succes! 🚀 Joburile partiale au fost salvate in: {PARTIAL_LOCATIONS_FILE}")
        
    except Exception as e:
        print(f"\nEroare la scrierea fisierului partial: {e}")
else:
    print("Info: Nu au fost gasite locatii partiale de revizuit.")
```

```python
import json
import os

# --- Configurari ---
PARTIAL_FILE = "locatii_partiale_de_revizuit.json"
MAIN_FILE = "E_Commers_jobs_cu_locatii_divizate.json"
FINAL_FILE = "E_Commers_jobs_done.json"

print("--- Incep procesul de finalizare ---")

# ======================================================================
# --- Partea 1: Corectarea fisierului partial ---
# ======================================================================
print(f"\n--- Partea 1: Actualizarea campului 'Location' in {PARTIAL_FILE} ---")

try:
    # Citim datele partiale
    with open(PARTIAL_FILE, 'r', encoding='utf-8') as f:
        partial_data = json.load(f)
    
    if not isinstance(partial_data, list):
        print(f"Eroare: {PARTIAL_FILE} nu contine o lista. Operatiune oprita.")
        exit()

    print(f"Am incarcat {len(partial_data)} joburi din {PARTIAL_FILE} pentru corectare.")
    
    # Corectam in memorie
    count_corectate = 0
    for job in partial_data:
        # Extragem valorile (chiar daca au spatii la inceput)
        city = job.get("city")
        region = job.get("region")
        country = job.get("country")
        
        # Construim lista de parti, curatandu-le de spatii extra
        parts = []
        if isinstance(city, str) and city.strip():
            parts.append(city.strip())
        if isinstance(region, str) and region.strip():
            parts.append(region.strip())
        if isinstance(country, str) and country.strip():
            parts.append(country.strip())
        
        # Cream noul string "Location" (ex: "Winnenden, Baden-Württemberg, Germany")
        new_location_str = ", ".join(parts)
        
        # Actualizam campul "Location" in job
        if job.get("Location") != new_location_str:
            job["Location"] = new_location_str
            count_corectate += 1

    print(f"Am actualizat campul 'Location' pentru {count_corectate} joburi.")

    # Suprascriem fisierul partial cu datele corectate
    with open(PARTIAL_FILE, 'w', encoding='utf-8') as f:
        json.dump(partial_data, f, indent=4, ensure_ascii=False)
    
    print(f"Succes! Fisierul {PARTIAL_FILE} a fost suprascris.")

except FileNotFoundError:
    print(f"Eroare: Fisierul {PARTIAL_FILE} nu a fost gasit. Opreste scriptul.")
    exit()
except json.JSONDecodeError:
    print(f"Eroare: Fisierul {PARTIAL_FILE} este corupt. Opreste scriptul.")
    exit()
except Exception as e:
    print(f"A aparut o eroare neasteptata in Partea 1: {e}")
    exit()

# ======================================================================
# --- Partea 2: Combinarea fisierelor ---
# ======================================================================
print(f"\n--- Partea 2: Combinarea {MAIN_FILE} si {PARTIAL_FILE} ---")

final_data_list = []

# Incarcam fisierul principal (cele completate automat cu Berlin)
try:
    with open(MAIN_FILE, 'r', encoding='utf-8') as f:
        main_data = json.load(f)
    
    if isinstance(main_data, list):
        final_data_list.extend(main_data)
        print(f"Am adaugat {len(main_data)} joburi din {MAIN_FILE}.")
    else:
        print(f"Avertisment: {MAIN_FILE} nu contine o lista. Va fi ignorat.")

except FileNotFoundError:
    print(f"Eroare: Fisierul {MAIN_FILE} nu a fost gasit. Opreste scriptul.")
    exit()
except Exception as e:
    print(f"Eroare la citirea {MAIN_FILE}: {e}")
    exit()

# Adaugam datele partiale (deja in memorie, acum corectate)
if partial_data:
    final_data_list.extend(partial_data)
    print(f"Am adaugat {len(partial_data)} joburi corectate din {PARTIAL_FILE}.")

print(f"\nTotal joburi in fisierul final: {len(final_data_list)}")

# Salvam fisierul final
try:
    with open(FINAL_FILE, 'w', encoding='utf-8') as f:
        json.dump(final_data_list, f, indent=4, ensure_ascii=False)
    
    print(f"\nSucces! 🚀 Toate datele au fost combinate si salvate in: {FINAL_FILE}")

except Exception as e:
    print(f"Eroare la scrierea fisierului final {FINAL_FILE}: {e}")
```
