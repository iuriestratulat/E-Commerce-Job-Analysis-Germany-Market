```python
import json

# Numele fisierelor
input_filename = 'E-Commerce_jobs_germany6month_19_10.json'
output_filename = 'extracted_jobs.json'

# Lista unde vom stoca datele extrase
extracted_data = []

try:
    # 1. Deschidem si citim fisierul JSON de intrare
    # Folosim encoding='utf-8' pentru a gestiona corect caracterele speciale (ex: germana)
    with open(input_filename, 'r', encoding='utf-8') as f_in:
        
        # Incarcam datele din fisier
        all_jobs_data = json.load(f_in)

        # Verificam structura datelor:
        # Presupunem ca fisierul contine o LISTA de joburi
        # Dar, bazat pe exemplul tau, ar putea fi si un singur obiect
        
        jobs_to_process = []
        if isinstance(all_jobs_data, list):
            # Cazul standard: fisierul contine o lista de joburi
            jobs_to_process = all_jobs_data
        elif isinstance(all_jobs_data, dict):
            # Cazul in care fisierul contine un singur job (nu o lista)
            jobs_to_process = [all_jobs_data]
        else:
            print(f"Eroare: Continutul JSON din '{input_filename}' nu este nici o lista, nici un obiect (dictionar).")

        # 2. Iteram prin fiecare job din lista de procesat
        for job in jobs_to_process:
            # Extragem titlul si descrierea
            # Folosim .get() pentru a evita erori daca un camp lipseste (va returna None)
            title = job.get('Title')
            description = job.get('Description')
            
            # Adaugam dictionarul nou in lista noastra
            extracted_data.append({
                'Title': title,
                'Description': description
            })

    # 3. Scriem datele extrase in fisierul JSON de iesire
    with open(output_filename, 'w', encoding='utf-8') as f_out:
        # json.dump() scrie lista in fisier
        # indent=4 formateaza fisierul JSON frumos (pretty-print)
        # ensure_ascii=False se asigura ca pastreaza caracterele speciale (ex: ü, ä, ö)
        json.dump(extracted_data, f_out, indent=4, ensure_ascii=False)

    print(f"Succes! Datele au fost extrase si salvate in '{output_filename}'.")

except FileNotFoundError:
    print(f"Eroare: Fisierul '{input_filename}' nu a fost gasit.")
except json.JSONDecodeError:
    print(f"Eroare: Fisierul '{input_filename}' nu este un JSON valid.")
except Exception as e:
    print(f"A aparut o eroare neasteptata: {e}")
```

```python
import json

# Numele fisierelor
input_filename = 'dataset_glassdoor_e_com.json'
output_filename = 'unique_skills.json'

# Folosim un 'set' pentru a colecta skill-urile.
# Un set stocheaza automat doar valori unice.
unique_skills_set = set()

try:
    # 1. Deschidem si citim fisierul JSON de intrare
    with open(input_filename, 'r', encoding='utf-8') as f_in:
        all_data = json.load(f_in)

        # Verificam daca datele sunt o lista (formatul standard pentru un JSON cu mai multe intrari)
        items_to_process = []
        if isinstance(all_data, list):
            items_to_process = all_data
        elif isinstance(all_data, dict):
            # Fallback in cazul in care fisierul contine un singur obiect, nu o lista
            items_to_process = [all_data]
        else:
            print(f"Eroare: Continutul JSON din '{input_filename}' nu este o lista sau un obiect.")
            exit() # Oprim scriptul daca formatul e gresit

        # 2. Iteram prin fiecare intrare (job/item)
        for item in items_to_process:
            # Extragem lista de skill-uri folosind .get() pentru a evita erori
            skills_list = item.get('skills')
            
            # Verificam daca 'skills_list' este intr-adevar o lista
            if isinstance(skills_list, list):
                # Adaugam fiecare skill din lista in set-ul nostru
                for skill in skills_list:
                    # Verificam daca skill-ul este un string si nu e gol
                    if isinstance(skill, str) and skill.strip():
                        unique_skills_set.add(skill.strip()) # .strip() curata spatiile albe

    # 3. Convertim set-ul intr-o lista sortata
    # Sortarea este optionala, dar face fisierul de iesire mai ordonat
    sorted_unique_skills = sorted(list(unique_skills_set))

    # 4. Cream dictionarul final pentru output, in formatul cerut
    output_data = {
        "skills": sorted_unique_skills
    }

    # 5. Scriem datele in fisierul JSON de iesire
    with open(output_filename, 'w', encoding='utf-8') as f_out:
        # indent=4 formateaza fisierul JSON frumos (pretty-print)
        # ensure_ascii=False pastreaza caracterele speciale
        json.dump(output_data, f_out, indent=4, ensure_ascii=False)

    print(f"Succes! Au fost extrase {len(sorted_unique_skills)} skill-uri unice.")
    print(f"Rezultatul a fost salvat in '{output_filename}'.")

except FileNotFoundError:
    print(f"Eroare: Fisierul '{input_filename}' nu a fost gasit.")
except json.JSONDecodeError:
    print(f"Eroare: Fisierul '{input_filename}' nu este un JSON valid.")
except Exception as e:
    print(f"A aparut o eroare neasteptata: {e}")
```

```python
import json

# Definim fisierele de intrare si iesire
input_files = ['unique_skills.json', 'json_skills1.json', 'json_skills2.json']
output_file = 'unique_skills_extended.json'

# Folosim un 'set' pentru a ne asigura ca toate skill-urile sunt unice
all_unique_skills = set()

# Parcurgem fiecare fisier de intrare
for filename in input_files:
    try:
        # 1. Deschidem si citim fisierul
        with open(filename, 'r', encoding='utf-8') as f_in:
            print(f"Procesare fișier: {filename}...")
            data = json.load(f_in)
            
            # 2. Extragem lista de sub cheia "skills"
            # Folosim .get("skills", []) pentru a returna o lista goala
            # daca cheia "skills" lipseste, evitand o eroare.
            skills_list = data.get('skills')

            # 3. Verificam daca am extras o lista valida
            if isinstance(skills_list, list):
                # Curatam si adaugam fiecare skill in set
                # .strip() elimina spatiile albe de la inceput/sfarsit
                cleaned_skills = {skill.strip() for skill in skills_list if isinstance(skill, str) and skill.strip()}
                
                # Adaugam skill-urile din lista curenta in set-ul principal
                all_unique_skills.update(cleaned_skills)
                print(f"  -> Adăugat {len(cleaned_skills)} skill-uri unice din {filename}.")
            else:
                print(f"  Avertisment: Fisierul '{filename}' nu are formatul asteptat (cheia 'skills' lipseste sau nu este o lista).")

    except FileNotFoundError:
        print(f"  Eroare: Fisierul '{filename}' nu a fost gasit. Se continuă...")
    except json.JSONDecodeError:
        print(f"  Eroare: Fisierul '{filename}' nu este un JSON valid. Se continuă...")
    except Exception as e:
        print(f"  A aparut o eroare neasteptata la citirea '{filename}': {e}. Se continuă...")

# 4. Convertim set-ul final intr-o lista si o sortam alfabetic
final_sorted_list = sorted(list(all_unique_skills))

# 5. Pregatim datele pentru fisierul de iesire
output_data = {
    "skills": final_sorted_list
}

# 6. Scriem rezultatul combinat in noul fisier JSON
try:
    with open(output_file, 'w', encoding='utf-8') as f_out:
        json.dump(output_data, f_out, indent=4, ensure_ascii=False)
    
    print(f"\nSucces!")
    print(f"Total skill-uri unice combinate: {len(final_sorted_list)}")
    print(f"Lista extinsă a fost salvată în '{output_file}'.")

except Exception as e:
    print(f"A aparut o eroare la scrierea fisierului de iesire '{output_file}': {e}")
```
